# 1. Dictionaryapi

```python
word = 'tomato'
response=requests.get('https://api.dictionaryapi.dev/api/v2/entries/en/'+word)
# response.status_code
```

```python
data = response.json()
# data
```

```python
definitions = []
for meaning in data[0]['meanings']:
    for definition in meaning['definitions']:
        definitions.append(definition['definition'])
```
